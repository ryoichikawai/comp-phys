```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton, bisect

def f(x):
    return x - np.tanh(beta*x)

def df(x):
    return 1 - beta/np.cosh(beta*x)**2

N=101
x0=2.0

t=np.linspace(0,2,N)+0.02
m=np.zeros(N)

for k in range(N):
    beta=1/t[k]

    m[k]=newton(f,x0,fprime=df,maxiter=200)


plt.ioff()
plt.figure(figsize=(12,5))
plt.plot(t,m, '-r')
plt.plot(t,-m, '-r')
plt.plot([0,1],[0,0],'--w')
plt.ylim([-1.2,1.2])
plt.xlabel('T')
plt.ylabel('m')
plt.savefig('ch04r05.pdf')
plt.show()
```
